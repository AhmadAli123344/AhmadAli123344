```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('Churn_Modelling.csv')
df=df.drop('RowNumber',axis=1)

plt.figure(1)
sns.countplot(data=df,x=df.Geography,hue='Gender')

plt.figure(2)
sns.countplot(data=df,x=df.Exited,hue='Gender')

from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Surname']=lab.fit_transform(df['Surname'])
df['Geography']=lab.fit_transform(df['Geography'])
df['Gender']=lab.fit_transform(df['Gender'])

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=df.drop('Exited',axis=1)
y=df['Exited']

X1=scaler.fit_transform(X)
print('Scaled X1:- \n',X1)
df
```

```python
import pickle
with open("Preprocess_variables.pkl", "wb") as f:
    pickle.dump((X1,y), f)
```
