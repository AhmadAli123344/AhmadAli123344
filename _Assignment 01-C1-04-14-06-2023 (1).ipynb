# Task 1

```python
import pandas as pd
import matplotlib.pyplot as plt
df1=pd.read_excel('Sample_Superstore.xls',sheet_name='Orders')
df1.fillna(0,inplace=True)
df1['Profit']=pd.to_numeric(df1['Profit'],errors='coerce')
Profit=df1.groupby('Region')['Profit'].sum()
Profit.plot(kind="bar")
plt.show()
```

# Task 2

```python
sale=df1.groupby('Sub-Category')['Sales'].sum()
sale.plot(kind="bar")
plt.show()
```

# Task 3

```python
df1['Quantity']=pd.to_numeric(df1['Quantity'],errors='coerce')
df1['Discount']=pd.to_numeric(df1['Discount'],errors='coerce')
plt.scatter(df1['Quantity'],df1['Discount'])
plt.xlabel('Quantity')
plt.ylabel('Discount')
plt.show()

```

# Task 4

```python
df1['Order Date']=pd.to_datetime(df1['Order Date'], errors='coerce')
monthly_sales = df1.groupby(df1['Order Date'].dt.month)['Sales'].sum()
plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trend')
plt.show()
```

# Task 5

```python
a=df1['Ship Mode'].value_counts()
a.plot(kind='pie',autopct='%1.1f%%')
plt.legend(title='Set of Data:')
plt.show()
```
