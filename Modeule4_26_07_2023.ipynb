```python
import pickle
with open("Preprocess_variables.pkl", "rb") as f:
    X1,y = pickle.load(f)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix

X_train,X_test,y_train,y_test=train_test_split(X1,y,test_size=0.2,random_state=42)

model=Sequential()
model.add(Dense(120,activation='sigmoid',input_shape=(12,)))
model.add(Dense(75,activation='sigmoid'))
model.add(Dense(45,activation='sigmoid'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

y_train_encoded = to_categorical(y_train)
print(y_train_encoded)

h=model.fit(X_train,y_train_encoded, batch_size=680, epochs=170, validation_split=0.2)
```

```python
import numpy as np
y_pred=model.predict(X_test)
y_pred_1 = np.argmax(y_pred, axis=1)

""" argmax('y_pred',axis=1) return maximum index location e.g[1 0 0] so argmax return 0 which means
    in that Vector which class is highly predict as we have [0 1 2] in y_test label of each class"""

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred_1)
print("Accuracy:", accuracy)

cf=confusion_matrix(y_test, y_pred_1)
print(cf)
```
