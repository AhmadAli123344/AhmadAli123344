# Scatterplots between all features

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('Cat_human_New.csv')
for col in df.columns:
    if col!='label':
        plt.figure(figsize=(5,5))
        plt.scatter(df['label'],df[col])
        plt.xlabel('label')
        plt.ylabel(col)
        plt.title(f'Scatter Graph for Cat vs Human - {col}')
        plt.grid()
        plt.show()
```

# Draw Line Graph

```python
# First Method

df.plot.line(legend=True)
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Graphs Cat vs Human')
plt.grid()
plt.show()
```

```python
# Seconad Method
for col in df.columns:
    if col!='label':
        plt.figure(figsize=(3,3))
        plt.plot(df['label'],df[col])
        plt.xlabel('label')
        plt.ylabel(col)
        plt.title(f'Scatter Graph for Cat vs Human - {col}')
        plt.grid()
        plt.show()
```

# Draw Histogram

```python
for col in df.columns:
    if col!='label':
        plt.figure(figsize=(8,8))
        plt.hist(df[col])
        plt.xlabel('label')
        plt.ylabel(col)
        plt.title(f'Hist Graph Cat vs Human - {col}')
        plt.grid()
        plt.show()
```

# Draw box plot

```python
df1=df.select_dtypes(include=['float64', 'int64'])
for column in df1.columns:
    plt.figure(figsize=(4,4))
    plt.boxplot(df1[column])
    plt.ylabel(column)
    plt.title(f'Box Graph for Cat vs Human - {column}')
    plt.show()
```

# Draw Bar Graph

```python
df1=df.select_dtypes(include=['float64', 'int64'])
for column in df1.columns:
    plt.figure(figsize=(4,4))
    plt.bar(df['label'],df1[column])
    plt.xlabel('label')
    plt.ylabel(column)
    plt.title(f'Bar Graph for Cat vs Human -{column}')
    plt.show()

```

# Draw Heat map

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(df1, cmap="YlGnBu")
plt.title('Heatmap of Numerical Data')
plt.show()
```
